```python
from collections import Counter
```

```python
def read_vocab(corpus_file):
    vocab = Counter()
    with open(corpus_file) as f:
        for line in f:
            tags, tokens, lc = get_tags_tokens_lowercase(line)
            vocab.update(tokens)
    return vocab
```
