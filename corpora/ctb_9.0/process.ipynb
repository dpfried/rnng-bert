```python
from nltk.corpus import BracketParseCorpusReader, LazyCorpusLoader
```

```python
ctb_9 = LazyCorpusLoader("ctb_9.0", BracketParseCorpusReader, r'chtb_.*', tagset='unknown')
```

```python
training = list(range(1,270+1)) + list(range(440,1151+1))
development = list(range(301, 325+1))
test = list(range(271,300+1))

others=list(range(1152, 7017+1))

# make sure there's no overlap
assert not (set(training) & set(development))
assert not (set(training) & set(test))
assert not (set(development) & set(test))
assert not (set(training) & set(others))
assert not (set(development) & set(others))
assert not (set(test) & set(others))

print("total num files: %d" % len(training + development + test))

process_ctb.combine_fids(ctb_9, nltk_out_root, training, 'train.gold.original', suffix_glob='*', pad_hundreds=True)
process_ctb.combine_fids(ctb_9, nltk_out_root, development, 'dev.gold.original', suffix_glob='*', pad_hundreds=True)
process_ctb.combine_fids(ctb_9, nltk_out_root, test, 'test.gold.original', suffix_glob='*', pad_hundreds=True)
process_ctb.combine_fids(ctb_9, nltk_out_root, others, 'others.gold.original', suffix_glob='*', pad_hundreds=True)
```
